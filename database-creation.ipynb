```python
import sqlite3
import pandas as pd
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE testee_table (Testee_id INTEGER PRIMARY KEY, Testee_title text, Testee_forename text,
            Testee_surname text, Testee_phone_number number, Testee_email text, Testee_postcode text, Test_date date)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE manufacturer_table (Manufacturer_id INTEGER PRIMARY KEY, Manufacturer text, 
            Kit_name text)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE supplier_table (Supplier_id INTEGER PRIMARY KEY, Supplier text, 
            Supplier_address text, Supplier_postcode text)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE primer_table (Primer_id INTEGER PRIMARY KEY, primer_name text, primer_sequence text, 
            primer_conc text, primer_supplier text, 
            primer_software text, primer_cost number)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE sequence_table (Sequence_id INTEGER PRIMARY KEY, Amplified_sequence text,
            Experiment_purpose text, Primer1_id INTEGER, Primer2_id INTEGER,
            FOREIGN KEY (Primer1_id) REFERENCES primer_table(Primer_id),
            FOREIGN KEY (Primer2_id) REFERENCES primer_table(Primer_id))""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE centre_table (Centre_id INTEGER PRIMARY KEY, Scientist_centre text, 
            Centre_postcode text, Area_code INTEGER, Subscriber_number INTEGER)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE scientist_table (Scientist_id INTEGER PRIMARY KEY, Scientist_title text, 
            Scientist_forename text, Scientist_surname text, Scientist_email text, Centre_id INTEGER,
            FOREIGN KEY (Centre_id) REFERENCES centre_table(Centre_id))""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE results_table (Results_id INTEGER PRIMARY KEY, Assessment text, Variant_name text, 
    Variant_sequence text, Results_date date)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE temperature_table (Temp_id INTEGER PRIMARY KEY, Denature_temp number, Annealing_temp number, 
    Elongate_temp number, Completion_temp number)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE time_table (Time_id INTEGER PRIMARY KEY, Denature_time time, Annealing_time time, 
    Elongate_time time, Completion_time time)""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE experiment_table (Experiment_id INTEGER PRIMARY KEY, Experiment_date date,
            Temp_id INTEGER, Time_id INTEGER, cycles INTEGER,
            FOREIGN KEY (Temp_id) REFERENCES temperature_table(Temp_id),
            FOREIGN KEY (Time_id) REFERENCES time_table (Time_id))""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE PCR_table (PCR_id INTEGER PRIMARY KEY, Order_number number,
            Manufacturer_id INTEGER, Supplier_id INTEGER, Sequence_id INTEGER, PCR_cost_pounds number, Buffer text, 
            Buffer_concentration_mol number, Enzyme text, Enzyme_concentration_mol number, Nucleotide_mix text, 
            nucleotide_concentration_mol number,
            FOREIGN KEY (Manufacturer_id) REFERENCES manufacturer_table(Manufacturer_id),
            FOREIGN KEY (Supplier_id) REFERENCES supplier_table (Supplier_id),
            FOREIGN KEY (Sequence_id) REFERENCES sequence_table (Sequence_id))""")
```

```python
with sqlite3.connect('my_database.db') as conn:
    c = conn.cursor()
    c.execute("""CREATE TABLE test_table (Test_id text, Testee_id INTEGER, PCR_id INTEGER, Scientist_id INTEGER,
    Experiment_id INTEGER, Results_id INTEGER,
            FOREIGN KEY (Testee_id) REFERENCES testee_table(Testee_id)
            FOREIGN KEY (PCR_id) REFERENCES PCR_table (PCR_id),
            FOREIGN KEY (Scientist_id) REFERENCES scientist_table (Scientist_id),
            FOREIGN KEY (Experiment_id) REFERENCES experiment_table (Experiment_id)
            FOREIGN KEY (Results_id) REFERENCES results_table (Results_id))""")
```
